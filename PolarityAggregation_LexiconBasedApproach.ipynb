```python
emotion_big_df.head()
```

```python
def polarity_weekly(year):
    big_df=pd.DataFrame()
    i=0
    polarity_avg=pd.DataFrame()
    polarity_count=pd.DataFrame()
    data=emotion_big_df[(emotion_big_df['createdAt']>=str(year)) & (emotion_big_df['createdAt']<str(year+1))]
    for i in range(1,53):
        df=data[data['date']==i]
        
        
        df_avg=df[['canton','Polarity']].groupby(['canton']).agg(['mean'])
        df_avg.columns = df_avg.columns.droplevel(level=1)
        df_avg.columns=[str(year)+'_'+str(i)+'_polarity']
        polarity_avg=pd.concat([polarity_avg,df_avg], axis=1)
                        
                        
                        
        df_count=df[['canton','Polarity']].groupby(['canton']).agg(['count'])
        df_count.columns = df_count.columns.droplevel(level=1)
        df_count.columns=[str(year)+'_'+str(i)+'_polarity']
        polarity_count=pd.concat([polarity_count,df_count], axis=1)


    return polarity_avg, polarity_count
```

```python
average_polarity_2012, count_polarity_2012=polarity_weekly(2012)
average_polarity_2013, count_polarity_2013=polarity_weekly(2013)
average_polarity_2014, count_polarity_2014=polarity_weekly(2014)
average_polarity_2015, count_polarity_2015=polarity_weekly(2015)
average_polarity_2016, count_polarity_2016=polarity_weekly(2016)
```

```python
average_polarity=pd.concat([average_polarity_2012,average_polarity_2013,average_polarity_2014
                           ,average_polarity_2015,average_polarity_2016], axis=1)
```

```python
count_polarity=pd.concat([count_polarity_2012,count_polarity_2013,count_polarity_2014
                           ,count_polarity_2015,count_polarity_2016], axis=1)
```

```python
average_polarity.reset_index(inplace=True)
count_polarity.reset_index(inplace=True)
```

```python
#weekdays

def weekdays_weekends(year):
    
    data=emotion_big_df[(emotion_big_df['createdAt']>=str(year)) & (emotion_big_df['createdAt']<str(year+1))]


    big_df=pd.DataFrame()
    i=0
    weekdays_avg=pd.DataFrame()
    weekdays_count=pd.DataFrame()

    weekends_avg=pd.DataFrame()
    weekends_count=pd.DataFrame()
    

    for i in range(1,53):
        df=data[data['date']==i]

        df_weekdays=df[(df['weekday']>=0) & (df['weekday']<5)]
        df_weekends=df[(df['weekday']==5) | (df['weekday']==6)]



        df_weekdays_avg=df_weekdays[['canton','Polarity']].groupby(['canton']).agg(['mean'])
        df_weekdays_avg.columns = df_weekdays_avg.columns.droplevel(level=1)
        df_weekdays_avg.columns=[str(year)+'_'+str(i)+'_polarity']
        weekdays_avg=pd.concat([weekdays_avg,df_weekdays_avg], axis=1)


        df_weekdays_count=df_weekdays[['canton','Polarity']].groupby(['canton']).agg(['count'])
        df_weekdays_count.columns = df_weekdays_count.columns.droplevel(level=1)
        df_weekdays_count.columns=[str(year)+'_'+str(i)+'_polarity']
        weekdays_count=pd.concat([weekdays_count,df_weekdays_count], axis=1)


        df_weekends_avg=df_weekends[['canton','Polarity']].groupby(['canton']).agg(['mean'])
        df_weekends_avg.columns = df_weekends_avg.columns.droplevel(level=1)
        df_weekends_avg.columns=[str(year)+'_'+str(i)+'_polarity']
        weekends_avg=pd.concat([weekends_avg,df_weekends_avg], axis=1)


        df_weekends_count=df_weekends[['canton','Polarity']].groupby(['canton']).agg(['count'])
        df_weekends_count.columns = df_weekends_count.columns.droplevel(level=1)
        df_weekends_count.columns=[str(year)+'_'+str(i)+'_polarity']
        weekends_count=pd.concat([weekends_count,df_weekends_count], axis=1)
        
        
    return weekdays_avg,weekdays_count,weekends_avg,weekends_count
        

```

```python
weekdays_avg_2012,weekdays_count_2012,weekends_avg_2012,weekends_count_2012=weekdays_weekends(2012)
weekdays_avg_2013,weekdays_count_2013,weekends_avg_2013,weekends_count_2013=weekdays_weekends(2013)
weekdays_avg_2014,weekdays_count_2014,weekends_avg_2014,weekends_count_2014=weekdays_weekends(2014)
weekdays_avg_2015,weekdays_count_2015,weekends_avg_2015,weekends_count_2015=weekdays_weekends(2015)
weekdays_avg_2016,weekdays_count_2016,weekends_avg_2016,weekends_count_2016=weekdays_weekends(2016)
```
