# V. Majority Voting:

```python
def find_majority(k):
    myMap = {}
    maximum = ( '', 0 ) # (occurring element, occurrences)
    for n in k:
        if n in myMap: myMap[n] += 1
        else: myMap[n] = 1

        # Keep track of maximum on the go
        if myMap[n] > maximum[1]: maximum = (n,myMap[n])

    return maximum
```

```python
majority_emotions = []
for i in range(0,len(lexicon_results_df)):
    votes = lexicon_results_df['Emotion'].iloc[i] + pmi_results_df['Emotion'].iloc[i] +word2vec_results_df['Emotion'].iloc[i]
    (majority,occurences) = find_majority(votes)
    majority_emotions.append(majority)
```
