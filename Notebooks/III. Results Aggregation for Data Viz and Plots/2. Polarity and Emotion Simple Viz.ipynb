# Emotion Visualization

```python
import matplotlib.pyplot as plt

def emotion_pieplot_single_month(canton, month,df):
    row=df.loc[df['canton'] == canton]
    
    needed_column=[]
    for column in row.columns.tolist():
        if(column.startswith(month)):
            needed_column.append(column)
    Anger=row[needed_column][month+'_Anger'].tolist()[0]
    Anticipation=row[needed_column][month+'_Anticipation'].tolist()[0]
    Disgust=row[needed_column][month+'_Disgust'].tolist()[0]
    Fear=row[needed_column][month+'_Fear'].tolist()[0]
    Joy=row[needed_column][month+'_Joy'].tolist()[0]
    Sadness=row[needed_column][month+'_Sadness'].tolist()[0]
    Surprise=row[needed_column][month+'_Surprise'].tolist()[0]
    Trust=row[needed_column][month+'_Trust'].tolist()[0]

    # Pie chart, where the slices will be ordered and plotted counter-clockwise:
    labels = 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise', 'Trust'

    sizes = [Anger, Anticipation, Disgust, Fear, Joy, Sadness, Surprise, Trust]
    
    max_ondex=sizes.index(max(sizes))
    explode=[0,0,0,0,0,0,0,0]
    explode[max_ondex]=0.2
    explode=tuple(explode)
    
    #explode = (0, 0, 0, 0,0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
    
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','blue','red','green','white']

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90, colors=colors)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.show()
```

```python
import matplotlib.pyplot as plt

def emotion_pieplot_single_season(canton, season,df):
    row=df.loc[df['canton'] == canton]
    
    needed_column=[]
    for column in row.columns.tolist():
        if(column.startswith(season)):
            needed_column.append(column)
    Anger=row[needed_column][season+'_Anger'].tolist()[0]
    Anticipation=row[needed_column][season+'_Anticipation'].tolist()[0]
    Disgust=row[needed_column][season+'_Disgust'].tolist()[0]
    Fear=row[needed_column][season+'_Fear'].tolist()[0]
    Joy=row[needed_column][season+'_Joy'].tolist()[0]
    Sadness=row[needed_column][season+'_Sadness'].tolist()[0]
    Surprise=row[needed_column][season+'_Surprise'].tolist()[0]
    Trust=row[needed_column][season+'_Trust'].tolist()[0]
 
    # Pie chart, where the slices will be ordered and plotted counter-clockwise:
    labels = 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise', 'Trust'

    sizes = [Anger, Anticipation, Disgust, Fear, Joy, Sadness, Surprise, Trust]
    
    max_ondex=sizes.index(max(sizes))
    explode=[0,0,0,0,0,0,0,0]
    explode[max_ondex]=0.2
    explode=tuple(explode)
        
    colors = ['darkcyan', 'gold', 'cyan', 'pink','violet','r','chartreuse','blue']

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90, colors=colors)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    plt.show()
```

```python
import matplotlib.pyplot as plt

def emotion_year_season_pieplot(canton, year,df):
    row=df.loc[df['canton'] == canton]
    
    needed_column=[]
    for column in row.columns.tolist():
        if(year in column):
            needed_column.append(column)
            
    Anger=[]
    Anticipation=[]
    Disgust=[]
    Fear=[]
    Joy=[]
    Sadness=[]
    Surprise=[]
    Trust=[]
    seasons=['winter','spring','summer','fall']
    for i in range(len(seasons)):
        season=seasons[i]
        Anger.append(row[needed_column][season+str(year)+'_Anger'].tolist()[0])
        Anticipation.append(row[needed_column][season+str(year)+'_Anticipation'].tolist()[0])
        Disgust.append(row[needed_column][season+str(year)+'_Disgust'].tolist()[0])
        Fear.append(row[needed_column][season+str(year)+'_Fear'].tolist()[0])
        Joy.append(row[needed_column][season+str(year)+'_Joy'].tolist()[0])
        Sadness.append(row[needed_column][season+str(year)+'_Sadness'].tolist()[0])
        Surprise.append(row[needed_column][season+str(year)+'_Surprise'].tolist()[0])
        Trust.append(row[needed_column][season+str(year)+'_Trust'].tolist()[0])
    

    # Pie chart, where the slices will be ordered and plotted counter-clockwise:
    labels = 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise', 'Trust'

    sizes1 = [Anger[0], Anticipation[0], Disgust[0], Fear[0], Joy[0], Sadness[0], Surprise[0], Trust[0]]
    sizes2 = [Anger[1], Anticipation[1], Disgust[1], Fear[1], Joy[1], Sadness[1], Surprise[1], Trust[1]]
    sizes3 = [Anger[2], Anticipation[2], Disgust[2], Fear[2], Joy[2], Sadness[2], Surprise[2], Trust[2]]
    sizes4 = [Anger[3], Anticipation[3], Disgust[3], Fear[3], Joy[3], Sadness[3], Surprise[3], Trust[3]]
    
    max_index1=sizes1.index(max(sizes1))
    explode1=[0,0,0,0,0,0,0,0]
    explode1[max_index1]=0.2
    explode1=tuple(explode1)
    
    max_index2=sizes2.index(max(sizes2))
    explode2=[0,0,0,0,0,0,0,0]
    explode2[max_index2]=0.2
    explode2=tuple(explode2)
    
    max_index3=sizes3.index(max(sizes3))
    explode3=[0,0,0,0,0,0,0,0]
    explode3[max_index3]=0.2
    explode3=tuple(explode3)
    
    max_index4=sizes4.index(max(sizes4))
    explode4=[0,0,0,0,0,0,0,0]
    explode4[max_index4]=0.2
    explode4=tuple(explode4)

    #explode = (0, 0, 0, 0,0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','blue','red','green','white']
   
    plt.figure(1, figsize=(10,10))
    
    plt.subplot(221)

    plt.pie(sizes1, explode=explode1, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90, colors=colors)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    plt.subplot(222)
    plt.pie(sizes2, explode=explode2, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90, colors=colors)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    plt.subplot(223)
    plt.pie(sizes3, explode=explode3, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90, colors=colors)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    plt.subplot(224)
    plt.pie(sizes4, explode=explode4, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90, colors=colors)
    
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    
    plt.subplots_adjust(top=1.22, bottom=0.08, left=0.10, right=1.75, hspace=0.35,
                    wspace=0.5)
   
    plt.show()
```

```python
season_data_mean=pd.read_csv('mean_emotio.csv')
```

```python
cantons=season_data_mean['canton'].tolist()
```

```python
season_data_mean.drop('canton', axis=1, inplace=True)
```

```python
def emotion_canton_year(year,emotion):
    needed_column=[]
    for i in season_data_mean.columns.tolist():
        if((i.split('_')[1]==emotion) & (i.split('_')[0][-1]==year)):
            needed_column.append(i)

    df=season_data_mean[needed_column]
    df['mean']=df.mean(axis=1)
    df['canton']=cantons
    max_index=df['mean'].argmax()
    #return df
    return(df.loc[max_index]['canton'])
```

```python
def stakked_plot_data(emotion,year):

    needed_column=[]
    for i in season_data_mean.columns.tolist():
        if((i.split('_')[1]==emotion) & (i.split('_')[0][-1]==year)):
            needed_column.append(i)

    #print(needed_column)
    df=season_data_mean[needed_column]
    df[emotion]=df.mean(axis=1)
    df['canton']=cantons    
    return df[[emotion]]   
```

```python
#positive
Anticipation=stakked_plot_data('Anticipation','4')
Joy=stakked_plot_data('Joy','4')
Surprise=stakked_plot_data('Surprise','4')
Trust=stakked_plot_data('Trust','4')
all_emotion_4=pd.concat([Anticipation,Joy,Surprise,Trust], axis=1)
```

```python
#positive
Anger=stakked_plot_data('Anger','4')
Disgust=stakked_plot_data('Disgust','4')
Fear=stakked_plot_data('Fear','4')
Sadness=stakked_plot_data('Sadness','4')
all_emotion_4=pd.concat([Anger,Disgust,Fear,Sadness], axis=1)
```

```python
data=all_emotion_4.as_matrix()
```

```python
data=data.transpose()
```

```python
width = 0.45
ind = np.arange(26) + 0.75

fig, ax = plt.subplots(1,1)
p0 = ax.bar(ind, data[0],  width, color = 'gold')#'cyan')
p1 = ax.bar(ind, data[1], width, bottom = data[0], color = 'violet')
p2 = ax.bar(ind, data[2], width, bottom = data[0] + data[1], color = 'chartreuse')
p3 = ax.bar(ind, data[3], width, bottom = data[0] + data[1] + data[2], color = 'blue')

ax.set_ylabel('Average Emotion')
ax.set_xlabel('canton')
ax.set_xticks (ind + width/2.)
ax.set_xticklabels( cantons, rotation = 70 )

fig.legend( (p0[0], p1[0], p2[0], p3[0]), ('Anticipation', 'Joy', 'Surprise', 'Trust') ,loc=2,prop={'size':9})
fig.tight_layout()
fig.show()
```

```python
def plot_polarity_cantons_monthly(canton_list):
    tt=polarity_monthly.T
    tt.polarity_monthly=tt.loc['canton']
    tt.columns=tt.loc['canton']
    tt.drop('canton',axis=0, inplace=True)
    tt.replace('No Data',0, inplace=True)
    tt=tt.head(12)
    mylist = list(range(len(tt)))

    ax=tt[canton_list].astype(float).plot(figsize=(20,9),xticks=mylist,fontsize = 20)
    

    ax.lines[-1].set_linewidth(5)
    ax.lines[-2].set_linewidth(5)
    ax.lines[-3].set_linewidth(5)
    ax.tick_params(size=10)

    labels=['JAN','FRB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
    plt.xticks(mylist, labels)
    
    plt.ylabel("Average Polarity Score",fontsize=20)
    plt.xlabel("Month",fontsize=20)
    plt.legend(loc=2,fontsize=18)
```
