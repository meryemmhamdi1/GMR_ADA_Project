# Creating Random Results File for Vizualization Purposes: to be replaced by the real results once we have them

```python
import pandas as pd
list_emotions = ['anger','anticipation','disgust','fear','joy','sadness','surprise','trust']
cantons = ['ZG', 'BL', 'SH', 'GE', 'UR', 'NE', 'GL', 'ZH', 'SG', 'JU', 'TI', 'SO', 'BE', 'LU', 'NW', 'BS', 'OW', 'TG', 'VD', 'VS', 'AG', 'AI', 'FR', 'AR', 'SZ', 'GR']
header = ['Canton', '01/2014-03/2014','04/2014-06/2014','07/2014-09/2014','10/2014-12/2014','01/2015-03/2015','04/2015-06/2015','07/2015-09/2015','10/2015-12/2015','01/2016-03/2016','04/2016-06/2016','07/2016-09/2016','10/2016-12/2016']
seasons = ['01/2014-03/2014','04/2014-06/2014','07/2014-09/2014','10/2014-12/2014','01/2015-03/2015','04/2015-06/2015','07/2015-09/2015','10/2015-12/2015','01/2016-03/2016','04/2016-06/2016','07/2016-09/2016','10/2016-12/2016']
random_viz = pd.DataFrame(columns = header)
random_viz['Canton'] = cantons
```

```python
import random
for i in range(0,len(random_viz)):
    for season in seasons:
        random_viz.iloc[i][season] = random.choice(list_emotions)
random_viz
random_viz.to_csv('RandomResults4Viz.csv',index=False)
```
