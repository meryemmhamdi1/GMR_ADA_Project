```python

def weekdays_weekends_emotion_count(year,option):
    
    big_df=pd.DataFrame()
    i=0
    single_emotion_df=pd.DataFrame()
    dat=emotion_big_df[(emotion_big_df['createdAt']>=str(year)) & (emotion_big_df['createdAt']<str(year+1))]

    for i in range(1,53):
        df=dat[dat['date']== int(i)]
        
        
        df_weekdays=df[(df['weekday']>=0) & (df['weekday']<5)]
        df_weekends=df[(df['weekday']==5) | (df['weekday']==6)]
        
        if(option=='weekdays'):
            df=df_weekdays[['canton','Anger']].groupby(['canton']).agg(['count'])
        
        
        elif(option == 'weekends'):
            df=df_weekends[['canton','Anger']].groupby(['canton']).agg(['count'])
                
        

        df.columns = df.columns.droplevel(level=1)
        df.columns=[str(year)+'_'+str(i)+'_tweets']
        
        big_df=pd.concat([big_df,df], axis=1)



    return big_df 
```
